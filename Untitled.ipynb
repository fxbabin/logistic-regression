```python
# df
def skewness(arr, mean, std):
    size = len(arr)
    coeff = size / ((size - 1) * (size - 2))
    skewness = 0
    for e in arr:
        skewness += ((e - mean) / std) ** 3
    return (coeff * skewness)

def kurtosis(arr, mean, std):
    n = len(arr)
    coeff = (n * (n + 1)) / ((n - 1) * (n - 2) * (n - 3))
    kurtosis = 0
    for e in arr:
        kurtosis += ((e - mean) / (std ** 2)) ** 4
    kurtosis -= (3 * ((((n - 1) ** 2)) / ((n - 2) * (n - 3))))
    return (kurtosis)
```

```python
import numpy as np

def get_quartile(arr, count, nb=1):
    res = arr[((count // 4) * nb)] + (( (arr[((count // 4) * nb) + 1] - arr[((count // 4) * nb)]) / 4) * nb)
    return (res)
    
def describe(df):
    desc = {}
    
    for e in ['name', 'count', 'mean', 'min', 'max', 'std', '25%', '50%', '75%', 'skewness']:
        desc[e] = []
    
#     count
#     min
#     max
    
    
    
    for e in df.columns:
#         print(df[e].iloc[0])

#         count
        if isinstance(df[e].iloc[0], float):
            desc['name'].append(e)
            count = 0
            summ = 0.0
            minn = 1.0e9
            maxx = -1.0e9
            arr = sorted(df[e][df[e].map(lambda x: str(x) != "nan")])
    
            for el in arr:
                if str(el).lower() == "nan":
                    continue
                if el < minn:
                    minn = el
                if el > maxx:
                    maxx = el
                count += 1
                summ += el
            
            mean = summ / count
            
            # variance
            var = 0.0
            for el in arr:
                var += ((el - mean) ** 2)
            var /= (count - 1)
            std = np.sqrt(var)
        
            desc['count'].append(count)
            desc['mean'].append(mean)
            desc['min'].append(minn)
            desc['max'].append(maxx)
            desc['std'].append(std)
            desc['25%'].append(get_quartile(arr, count, 1))
            desc['50%'].append(get_quartile(arr, count, 2))
            desc['75%'].append(get_quartile(arr, count, 3))
            desc['skewness'].append(skewness(arr, mean, std))

    for e in ['name', 'count', 'mean', 'min', 'max', 'std', '25%', '50%', '75%', 'skewness']:
        out = ""
        if e == "name":
            out += "\t\t\t"
        else:
            out += e + "\t\t\t"
        
        for el in desc[e]:
            out += str("{}".format(el)) + "\t"
        out += "\n"
        print(out)
    return desc
d = describe(df)
```
